```python
import pandas as pd #PREPROCESSING

# Re-loading the dataset due to execution state reset
file_path = "C:\\Users\\Parag\\Desktop\\assesment2\\parag\\Crime_Data_from_2010_to_2019.csv"
crime_data = pd.read_csv(file_path)

# Preprocessing Steps

# 1. Handling Missing Values
# Filling missing values in 'Vict Sex' and 'Vict Descent' with 'Unknown'
# Dropping columns with a large number of missing values that are not critical for basic analysis
columns_to_drop = ['Mocodes', 'Weapon Used Cd', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']
crime_data['Vict Sex'].fillna('Unknown', inplace=True)
crime_data['Vict Descent'].fillna('Unknown', inplace=True)
crime_data.drop(columns=columns_to_drop, inplace=True)

# 2. Convert Data Types
# Converting 'Date Rptd' and 'DATE OCC' to datetime format
crime_data['Date Rptd'] = pd.to_datetime(crime_data['Date Rptd'])
crime_data['DATE OCC'] = pd.to_datetime(crime_data['DATE OCC'])

# 3. Handle Anomalous Values
# Replacing 0 values in LAT and LON with NaN
crime_data['LAT'].replace(0, pd.NA, inplace=True)
crime_data['LON'].replace(0, pd.NA, inplace=True)

# 4. Data Transformation
# Extracting year and month from 'DATE OCC' for easier analysis
crime_data['Year'] = crime_data['DATE OCC'].dt.year
crime_data['Month'] = crime_data['DATE OCC'].dt.month

# Checking the dataset after preprocessing
crime_data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Re-loading the dataset due to execution state reset
file_path = "C:\\Users\\Parag\\Desktop\\assesment2\\parag\\Crime_Data_from_2010_to_2019.csv"
crime_data = pd.read_csv(file_path)

# Preprocessing the dataset again quickly for visualization
crime_data['Date Rptd'] = pd.to_datetime(crime_data['Date Rptd'])
crime_data['DATE OCC'] = pd.to_datetime(crime_data['DATE OCC'])
crime_data['Year'] = crime_data['DATE OCC'].dt.year
crime_data['Month'] = crime_data['DATE OCC'].dt.month

# Setting up the visualization style
sns.set(style="whitegrid")

# Exploratory Data Analysis and Visualization

# 1. Yearly Crime Trends
yearly_crime_counts = crime_data['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
yearly_crime_counts.plot(kind='bar', color='skyblue')
plt.title('Yearly Crime Counts (2010-2019)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()

# 2. Crime Type Distributions
top_crime_types = crime_data['Crm Cd Desc'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_crime_types.values, y=top_crime_types.index, palette='viridis')
plt.title('Top 10 Crime Types')
plt.xlabel('Number of Occurrences')
plt.ylabel('Crime Type')
plt.show()

```

```python
# Monthly Crime Trends Analysis
monthly_crime_counts = crime_data.groupby('Month').size()

# Plotting Monthly Crime Trends
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_crime_counts.index, y=monthly_crime_counts.values)
plt.title('Monthly Crime Trends')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()

# Time of Day Analysis
# Extracting hour from TIME OCC
crime_data['Hour'] = crime_data['TIME OCC'] // 100
hourly_crime_counts = crime_data.groupby('Hour').size()

# Plotting Time of Day Crime Trends
plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_crime_counts.index, y=hourly_crime_counts.values)
plt.title('Time of Day Crime Trends')
plt.xlabel('Hour of Day (24-hour format)')
plt.ylabel('Number of Crimes')
plt.tight_layout()
plt.show()

# Geographical Analysis - Area wise Crime Distribution
area_crime_counts = crime_data['AREA NAME'].value_counts()

# Plotting Area wise Crime Distribution
plt.figure(figsize=(10, 6))
sns.barplot(y=area_crime_counts.index, x=area_crime_counts.values)
plt.title('Area Wise Crime Distribution')
plt.xlabel('Number of Crimes')
plt.ylabel('Area Name')
plt.tight_layout()
plt.show()

# Correlation Analysis - To be done later as it requires more complex statistical methods

# Type of Crime by Year
crime_type_by_year = pd.crosstab(crime_data['Year'], crime_data['Crm Cd Desc'])
top_crimes = crime_data['Crm Cd Desc'].value_counts().head(5).index
crime_type_by_year_top = crime_type_by_year[top_crimes]

# Plotting Type of Crime by Year for Top 5 Crimes
plt.figure(figsize=(10, 6))
sns.lineplot(data=crime_type_by_year_top)
plt.title('Top 5 Types of Crime by Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Type', labels=top_crimes)
plt.tight_layout()
plt.show()
```
